```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('student_dataset.csv')
```

### Exploratory Data Analysis
Plot and display the histograms of the following: G1, G2, and G3. Select gender as hues.

```python
# create subplots
fig, (ax1, ax2, ax3) = plt.subplots(nrows=3,ncols=1, figsize=(7,10))

# plot histograms
sns.histplot(data=data, x='G1', hue='sex', ax=ax1, multiple='stack')
sns.histplot(data=data, x='G2', hue='sex', ax=ax2, multiple='stack')
sns.histplot(data=data, x='G3', hue='sex', ax=ax3, multiple='stack')

# set labels
ax1.set(xlabel='first period grade')
ax2.set(xlabel='second period grade')
ax3.set(xlabel='final grade')
```
